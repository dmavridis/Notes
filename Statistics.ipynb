**Goodness of Fit Test** is used to perform hypothesis tests to copmare two or more populations

Why not t-test? - Chi-square is more approprate for evaluating data sets in which data is categorized. With t-test a hypothesis is evaluated when two sets of data are collected but not categorized. 

## Curves and distribution

Alternative Hypothesis $H_a$ = The stated distribution is not accurate 

**Calculate Chi-Square Statistic**

$$X^2  = \Sigma\frac{(observed - expected)^2}{expected}$$ 

The smaller the value the better the goodness of fit

# ANOVA: Analysis of Variance
## Introduction
Checks whether the variation of a particular output is a result of a factor of randomness.

### ANOVA Assumptions
- Normally distributed
- Independent observation
- Equivalent variance

## One-way ANOVA and the total sum of squares (SST)
Sum of squares of each point to grand mean

## Variance within and variance between (SSW and SSB)
Sum of squares of each point to category mean

SSW: Sum of square of each category mean to grand mean times number of elements in each category

$$\boxed{SST = SSW + SSB}$$

## Hypothesis testing and F-statistic
Null hypothesis is always the same. 

$H_0$ = Population means are equal

$H_a$ = Population means are not equal

$$F-Stat = \frac{\frac{SSB}{m-1}}{\frac{SSW}{n_t-m}}$$

Big F-statistic = Big difference in the categories (reject $H_0$)
